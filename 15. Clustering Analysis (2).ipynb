### Data Standardization and Encoding

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# Features (drop ID# and target)
X = df.loc[:, ~df.columns.isin(['ID#', 'Award?'])]

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
silhouette_k = silhouette_score(X_scaled, kmeans_labels)
print(f"K-Means Silhouette Score: {silhouette_k:.3f}")

# DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)
if len(set(db_labels)) > 1:
    silhouette_db = silhouette_score(X_scaled, db_labels)
    print(f"DBSCAN Silhouette Score: {silhouette_db:.3f}")

# Assign to dataframe
df['KMeans_Cluster'] = kmeans_labels
df['DBSCAN_Cluster'] = db_labels

print(df.head())
```

```python
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# K-MEANS CLUSTERING
# Finding optimal K using Elbow method
sse = []
K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# Plot Elbow curve
plt.figure(figsize=(6,4))
plt.plot(K_range, sse, 'o-', color='blue')
plt.xlabel('Number of clusters (K)')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal K')
plt.show()

# Fit K-Means with chosen K 
k_opt = 4
kmeans = KMeans(n_clusters=k_opt, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Silhouette Score
silhouette_k = silhouette_score(X_scaled, kmeans_labels)
print(f"K-Means Silhouette Score: {silhouette_k:.3f}")
```

```python
# Visualize clustering results
# Reduce to 2D for visualization using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(12,5))

# K-Means visualization
plt.subplot(1,2,1)
plt.scatter(X_pca[:,0], X_pca[:,1], c=kmeans_labels, cmap='tab10')
plt.title('K-Means Clustering')
plt.xlabel('PCA1')
plt.ylabel('PCA2')

# DBSCAN visualization
plt.subplot(1,2,2)
plt.scatter(X_pca[:,0], X_pca[:,1], c=db_labels, cmap='tab10')
plt.title('DBSCAN Clustering')
plt.xlabel('PCA1')
plt.ylabel('PCA2')

plt.show()
```

```python

# Cluster analysis (insights)
df['KMeans_Cluster'] = kmeans_labels
df['DBSCAN_Cluster'] = db_labels

# Example insights (print cluster stats)
print("K-Means Cluster Statistics:")
print(df.groupby('KMeans_Cluster').mean())

print("\nDBSCAN Cluster Statistics:")
print(df.groupby('DBSCAN_Cluster').mean())
```

```python
#Different Parameter Combinations
#epsilon ()  neighborhood radius

#minPts  minimum points to form a dense region
from sklearn.cluster import DBSCAN
import numpy as np

eps_values = [0.3, 0.5, 0.7, 1.0]
min_samples_values = [3, 5, 10]

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)

        # Count clusters (excluding noise)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)

        if n_clusters > 1:
            score = silhouette_score(X_scaled[labels != -1], labels[labels != -1])
        else:
            score = "Not defined"

        print(f"eps={eps}, minPts={min_samples}, clusters={n_clusters}, noise={n_noise}, silhouette={score}")
```
